```python
#1
class hinhtru():
  def __init__(self,bankinh,chieucao):
    self.bankinh=bankinh
    self.chieucao=chieucao
  def dientichbemat(self):
    return 2*3.14*self.bankinh*self.chieucao
  def thetich(self):
    return 3.14*self.bankinh*self.bankinh*self.chieucao
r,h=map(float,input().split())
kq=hinhtru(r,h)
print('%.2f'%(kq.dientichbemat()),'%.2f'%(kq.thetich()),sep=' ')
```

```python
#3
from math import *
class point():
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def check(self):
    if self.x==0 and self.y==0:
      return 1
    elif self.x==0:
      return 2
    elif self.y==0:
      return 3
  def dis(self):
    return '%.2f'%(sqrt(self.x*self.x+self.y*self.y))
x,y=map(float,input().split())
poi=point(x,y)
if poi.check()==1:
  print("M la goc toa do")
elif poi.check()==2:
  print("M nam tren truc tung")
elif poi.check()==3:
  print("M nam tren truc hoanh")
print(poi.dis())

```

```python
#5
class shop():
  def __init__(self,ds=[]):
    self.ds=ds
  def them(self,ten,gia):
    self.ds.append([ten,gia])
  def xoa(self,ten):
    for i in range(len(self.ds)):
      if self.ds[i][0]==ten:
        self.ds.remove(self.ds[i])
        break
  def check(self):
    if len(self.ds)==0:
      return False
    return True
  def sum(self):
    tong=0
    for i in range(len(self.ds)):
      tong+=self.ds[i][1]
    return tong
  def hienthi(self):
    for i in range(len(self.ds)):
      print(self.ds[i][0],self.ds[i][1])
  def xoatoanbo(self):
    self.ds.clear()
ds=shop()
while True:
  s=input('Có muốn thoát khỏi chương trình không?')
  if s=='yes':
    break
  id=int(input())
  if id==2:
    a=input()
    b=int(input())
    ds.them(a,b)
  elif id==3:
    a=input()
    ds.xoa(a)
  elif id==4:
    print(ds.check())
  elif id==5:
    print(ds.sum())
  elif id==6:
    print(ds.hienthi())
  elif id==7:
    print(ds.xoatoanbo())


```

```python
#6
class retangle():
  def __init__(self,w,h):
    self.w=w
    self.h=h
  def cv(self):
    return '%.2f'%((self.w+self.h)*2)
  def dt(self):
    return '%.2f'%((self.w*self.h))
  def scale(self,k):
    self.w*=k
    self.h*=k
w,h,k=map(float,input().split())
hcn=retangle(w,h)
hcn.scale(k)
print(hcn.cv(),hcn.dt())
```

```python
#7
from math import *
class phanso():
  def __init__(self,tu1,mau1,tu2,mau2):
    self.tu1=tu1
    self.mau1=mau1
    self.tu2=tu2
    self.mau2=mau2
  def cong(self):
    x=self.tu1*self.mau2+self.tu2*self.mau1
    y=self.mau1*self.mau2
    while gcd(x,y) != 1:
      u=gcd(x,y)
      x//=u
      y//=u
    print(x,'/',y,sep='')
  def tru(self):
    x=self.tu1*self.mau2-self.tu2*self.mau1
    y=self.mau1*self.mau2
    while gcd(x,y) != 1:
      u=gcd(x,y)
      x//=u
      y//=u
    print(x,'/',y,sep='')
  def nhan(self):
    x=self.tu1*self.tu2
    y=self.mau1*self.mau2
    while gcd(x,y) != 1:
      u=gcd(x,y)
      x//=u
      y//=u
    print(x,'/',y,sep='')
  def chia(self):
    x=self.tu1*self.mau2
    y=self.mau1*self.tu2
    while gcd(x,y) != 1:
      u=gcd(x,y)
      x//=u
      y//=u
    print(x,'/',y,sep='')
a=phanso(2,4,-1,4)
a.chia()


```

```python
#8
class bank():
  def __init__(self,ten,sodu):
    self.ten=ten
    self.sodu=sodu
  def rut(self,x):
    if self.sodu >= x:
      self.sodu-=x
  def them(self,x):
    self.sodu+=x
ten=input()
sodu=int(input())
a=bank(ten,sodu)
n=int(input())
while n>0:
  n-=1
  s=input()
  x=int(input())
  if s=='DEPOSIT':
    a.them(x)
  else:
    a.rut(x)
print('%.2f'%(a.sodu))
```

```python
#10
class Book:
    def __init__(self,title,author,year):
        self.title = title
        self.author = author
        self.year = year
class Library:
    def __init__(self):
        self.book = []
    def add(self,book):
        self.book.append(book)
    def count_by_author(self,name):
        cnt = 0
        for book in self.book:
            if book.author == name:
                cnt += 1
        return cnt
    def find_by_year(self,year):
        cnt = 0
        for book in self.book:
            if book.year == year:
                cnt += 1
        return cnt
n = int(input())
lib = Library()
while n > 0:
    thao_tac = input().split()
    if thao_tac[0] == 'ADD':
        b = thao_tac[1].split(';')
        book = Book(b[0],b[1],b[2])
        lib.add(book)
    elif thao_tac[0] == 'COUNT':
        print(lib.count_by_author(thao_tac[1]))
    elif thao_tac[0] == 'COUNTYEAR':
        print(lib.find_by_year(thao_tac[1]))
    n -= 1
```

```python
#11
from math import *
class Angle :
    def __init__(self,a,b,c,d):
        self.ABx = b[0] - a[0]
        self.ABy = b[1] - a[1]
        self.ABz = b[2] - a[2]
        self.ACx = c[0] - a[0]
        self.ACy = c[1] - a[1]
        self.ACz = c[2] - a[2]
        self.BCx = c[0] - b[0]
        self.BCy = c[1] - b[1]
        self.BCz = c[2] - b[2]
        self.BDx = d[0] - b[0]
        self.BDy = d[1] - b[1]
        self.BDz = d[2] - b[2]
    def vector(self):
        self.ABC = [self.ABy * self.ACz - self.ABz * self.ACy, self.ABz * self.ACx - self.ABx * self.ACz, self.ABx * self.ACy - self.ABy * self.ACx]
        self.BCD = [self.BCy * self.BDz - self.BCz * self.BDy, self.BCz * self.BDx - self.BCx * self.BDz, self.BCx * self.BDy - self.BCy * self.BDx]
    def cos(self):
        tich_vo_huong = self.ABC[0] * self.BCD[0] + self.ABC[1] * self.BCD[1] + self.ABC[2] * self.BCD[2]
        tich_do_dai = ((self.ABC[0] ** 2 + self.ABC[1] ** 2 + self.ABC[2] ** 2) ** 0.5) *((self.BCD[0] ** 2 + self.BCD[1] ** 2 + self.BCD[2] ** 2) ** 0.5)
        cos = abs(tich_vo_huong/tich_do_dai)
        return cos
A = list(map(float,input().split()))
B = list(map(float,input().split()))
C = list(map(float,input().split()))
D = list(map(float,input().split()))
my_angle = Angle(A,B,C,D)
my_angle.vector()
my_angle = acos(my_angle.cos())
my_angle = degrees(my_angle)
print(f'{my_angle:.2f}')
```

```python
#12
class Complex:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def pro(self,other):
        self.thuc = self.a * other.a - self.b * other.b
        self.ao = self.a * other.b + self.b * other.a
a = list(map(int,input().split()))
b = list(map(int,input().split()))
my_complex = Complex(a[0],a[1])
other_complex = Complex(b[0],b[1])
my_complex.pro(other_complex)
print(my_complex.thuc,my_complex.ao)
```

```python
#13
class Triangle:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def getArea(self):
        if self.a + self.b <= self.c or self.b + self.c <= self.a or self.a + self.c <= self.b:
            return -1
        self.p = (self.a + self.b + self.c)/2
        return (self.p * (self.p -self.a) * (self.p - self.b) * (self.p -self.c)) ** 0.5
my_tri = Triangle(*map(int,input().split()))
if my_tri.getArea() == -1 :
    print('invalid')
else:
    print(f'{my_tri.getArea():.2f}')
```
