```python
#1
def sapxep(a,b):
  kq=[]
  while len(a) > 0 and len(b) > 0:
    if a[0] < b[0]:
      kq.append(a[0])
      a.pop(0)
    else:
      kq.append(b[0])
      b.pop(0)
  if len(a) > 0:
    while len(a) > 0:
      kq.append(a[0])
      a.pop(0)
  elif len(b) > 0:
    while len(b) > 0:
      kq.append(b[0])
      b.pop(0)
  return kq
a=list(map(int,input().split()))
b=list(map(int,input().split()))
print(sapxep(a,b))
```

```python
#3
def swap(n):
    min = n[0]
    for i in range(len(n)):
        if int(n[i]) <= int(min):
            min = n[i]
            vi_tri = i
    if min == n[0]:
        n[-1],n[-2] = n[-2],n[-1]
    else:
        n[0],n[vi_tri] = n[vi_tri],n[0]
        if n[0] == '0': n.remove(n[0])
    return n
n = input()
n = list(n)
m = swap(n)
print(''.join(m))


```

```python
#6
n=int(input())
a=list(map(int,input().split()))
id=-1
for i in range(n-2,-1,-1):
  if a[i] < a[i+1]:
    id=i
    break
for i in range(n-1,id,-1):
  if a[i] > a[id]:
    a[i],a[id]=a[id],a[i]
    break
l=id+1
r=n-1
while l < r:
  a[l],a[r]=a[r],a[l]
  l+=1
  r-=1
print(a)
```

```python
#8
def check(s):
  l=[[0,0]]
  x=0
  y=0
  for i in range(len(s)):
    if s[i]=='R':
      x+=1
      if [x,y] in l:
        return True
      else:
        l.append([x,y])
    elif s[i]=='L':
      x-=1
      if [x,y] in l:
        return True
      else:
        l.append([x,y])
    elif s[i]=='U':
      y+=1
      if [x,y] in l:
        return True
      else:
        l.append([x,y])
    elif s[i]=='D':
      y-=1
      if [x,y] in l:
        return True
      else:
        l.append([x,y])
  return False
s=input()
print(check(s))
```
