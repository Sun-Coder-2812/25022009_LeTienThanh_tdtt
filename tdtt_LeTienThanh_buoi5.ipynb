```python
def Max_of_two(a,b):
  if a > b:
    return a
  else:
    return b
print(Max_of_two(5,9))
```

```python
def swap(a,b):
  a,b=b,a
  return [a,b]
print(swap(5,9))
```

```python
from math import *
def snt(n):
  for i in range(2,int(sqrt(n))+1):
    if n%i==0:
      return False
  return True
print(snt(6))
```

```python
from math import *
def perfect(n):
  sum=0
  for i in range(1,n):
    if n%i==0:
      sum+=i
  if sum == n:
    return True
  return False
print(perfect(123))
```

```python
def lis(a,k):
  for i in range(len(a)):
    if a[i]==k:
      return i
  return -1
a=[1,2,3,4,5]
k=6
print(lis(a,k))
```

```python
def giaithua(n):
  if n==0:
    return 1
  kq=1
  for i in range(1,n+1):
    kq*=i
  return kq
print(giaithua(10))
```

```python
def casio(s):
  if '+' in s:
     a=s.split('+')
     return ('%.2f'%(float(a[0])+float(a[1])))
  elif '-' in s:
     a=s.split('-')
     return ('%.2f'%(float(a[0])-float(a[1])))
  elif '*' in s:
     a=s.split('*')
     return ('%.2f'%(float(a[0])*float(a[1])))
  elif '/' in s:
     a=s.split('/')
     return ('%.2f'%(float(a[0])/float(a[1])))
s=input()
print(casio(s))
```

```python
def haming(a,b):
  if a > b:
    a,b=b,a
  bina=str(bin(a))
  binb=str(bin(b))
  stra=[]
  strb=[]
  for i in range(len(bina)-1,1,-1):
    stra.append(bina[i])
  for i in range(len(binb)-1,1,-1):
    strb.append(binb[i])
  while len(stra) < len(strb):
    stra.append('0')
  kq=0
  for i in range(len(stra)):
    if stra[i]!=strb[i]:
      kq+=1
  return kq
a,b=map(int,input().split())
print(haming(a,b))

```

```python
def sum(n):
  kq=0
  while n>0:
    kq+=(n%10)
    n//=10
  return kq
print(sum(12345))
```
