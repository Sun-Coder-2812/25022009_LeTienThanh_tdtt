```python
#1
def unique_sorted_student_ids(lst1, lst2) :
    merge = lst1 + lst2
    merge = [int(n) for n in merge]
    res = list(sorted(set(merge)))
    res = [str(n) for n in res]
    return res
assert unique_sorted_student_ids([], []) == []
assert unique_sorted_student_ids(["25020001"], []) == ["25020001"]
assert unique_sorted_student_ids(["25020003", "25020001", "25020001"], ["25020002"]) == ["25020001", "25020002", "25020003"]
assert unique_sorted_student_ids([25020005, "25020004"], ["25020004", 25020006]) == ["25020004", "25020005", "25020006"]
assert unique_sorted_student_ids(["10", "2"], ["3"]) == ["2", "3", "10"]
print("Problem 1 tests passed!")
```

```python
#2
def top3_frequent_names(a):
    dct = {}
    for i in a:
        if dct.get(i, 0) == 0:
            dct[i] = 1
        else:
            dct[i] += 1
    dct = dict(sorted(dct.items(), key = lambda x: x[1]))
    res = []
    for i in dct.keys():
        res.append(i)
    return res[::-1]
```

```python
#3
def most_frequent_triple(a):
    dct = {}
    res = ()
    c = 0
    for i in a:
        if dct.get(i,0) == 0:
            dct[i] = 1
        else:
            dct[i] += 1
        if dct[i] > c:
            c = dct[i]
            res = i
    return res
```

```python
#4
def reconcile_users(a,b):
    arr1 = []
    arr2 = []
    arr3 = []
    dct = {}
    for i in b:
        if i not in a and dct.get(i,0) == 0:
            arr1.append(i)
            dct[i] = 1
    for i in a:
        if i not in b and dct.get(i,0) == 0:
            dct[i] = 1
            arr2.append(i)
    for i in a:
        if i in b:
            arr3.append(i)
    return (arr1, arr2, arr3)
```

```python
#5
def contains_duplicate(a):
    dct = {}
    for i in a:
        if dct.get(i,0) == 0:
            dct[i] = 1
        else:
            return True
    return False
```

```python
#6

def check(a):
    return a % 3 == 0 and a%6 != 0

def squares_div_by3_not6(a):
    a = [int(n * n) for n in a if check(n)]
    return a
```

```python
#7
def even_numbers_in_matrix(a):
    a = [int(n) for row in a for n in row if n % 2 == 0]
    return a
```

```python
#8
def flatten_2d(a):
    return [n for row in a for n in row]
```

```python
#9
def contains_nearby_duplicate(nums, k):
    if k <= 0:return False
    for i in range(0,len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] == nums[j] and abs(i-j) <=k:
                return True
    return False
```

```python
#10
def merge_similar_items(items1, items2):
    d1 = {}
    d2 = {}
    for item in items1:
        d1[item[0]] = item[1]
    for item in items2:
        d2[item[0]] = item[1]
    #print(d1)
    #print(d2)
    for key in d1.keys():
        if key in d2.keys():
            d1[key] += d2[key]
    #print(d1)
    for key in d2.keys():
        if key not in d1.keys():
            d1[key] = d2[key]
    tem = [[key,val] for key, val in d1.items()]
    tem.sort()
    return tem
merge_similar_items([[1,1],[4,5],[3,8]], [[3,1],[1,5]])
merge_similar_items([[1,3],[2,2]], [[7,1],[2,2],[1,4]])
```

```python
#11
def farm_configurations(n):
    if n % 2 != 0:
        return []
    res = []
    for i in range(0, n//2 + 1):
        if (n - (2*i)) % 4 == 0:
            res.append((i, (n-2*i)//4))
    res.sort(reverse=True)
    return res
farm_configurations(4)
```

```python
#12
def can_make_odd_sum(a):
    odd_check = False
    even_check = False
    for i in a :
        if i % 2 != 0:
            odd_check = True
        elif i % 2 != 0:
            even_check = True
    return (odd_check and even_check) or (odd_check and len(a) % 2 != 0)
```
