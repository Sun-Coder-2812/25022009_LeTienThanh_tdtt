```python
def isTria(a,b,c):
    if a + b > c and a + c > b and b + c > a: return True
    return False 

a,b,c = map(int,input().split())
if not isTria(a,b,c) : print("Không phải tam giác") 
else:
    if a == b and b == c: print("Tam giác đều")
    elif a == b and b != c or a == c and b != a or b == c and a!=c: print("Tam giác cân")
    else: print("Tam giác thường")
```

```python
# ROUNDING
a = float(input())
phannguyen = int(a)

if a >= 0: 
    val3 = phannguyen
else: 
    if a == phannguyen: 
        val3 = phannguyen
    else: 
        val3 = phannguyen - 1 

if a >= 0:
    if a == phannguyen: 
        val1 = phannguyen
    else: 
        val1 = phannguyen + 1
else: 
    val1 = phannguyen

digit = a - phannguyen
if digit >= 0.5: 
    val2 = phannguyen + 1
elif digit <= -0.5: 
    val2 = phannguyen - 1
else: 
    val2 = phannguyen

print(val1, val2, val3)
```
